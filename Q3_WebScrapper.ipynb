```python
from bs4 import BeautifulSoup
import requests
import random
import time
import csv

# url='https://www.adityabirlacapital.com/healthinsurance/locate-care/hospital-listing'
url='https://general.bajajallianz.com/BagicNxt/hm/hmSearchExternalHosp.do'
req=requests.get(url)
content=req.text
soup=BeautifulSoup(content)
s=soup.select('title')
print(s)
print(s[0])
s[0].get_text()
maindiv=soup.find('div',class_='maincontent')
state=maindiv.p.text
print(state)
```
